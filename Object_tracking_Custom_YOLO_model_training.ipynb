```python
from ultralytics import YOLO
import os

# --- Training Hyperparameters ---
CONFIG_FILE = 'racket_config.yaml'
MODEL_WEIGHTS = 'yolov8n.pt' #'bat_tracking_v2.pt' #'yolov8n.pt'  # Nano model (fast)
RUN_NAME = 'yolov8n_racket_detector'
NUM_EPOCHS = 50
IMAGE_SIZE = 640
BATCH_SIZE = 16
PROJECT_DIR = 'runs/detect' # Specify the parent folder for results

print(f"Starting YOLOv8 training for {NUM_EPOCHS} epochs...")

try:
    # Load a pre-trained model for transfer learning
    model = YOLO(MODEL_WEIGHTS)

    # Start training
    # The model.train() function automatically saves checkpoints and the final 'best.pt' file.
    results = model.train(
        data=CONFIG_FILE,
        epochs=NUM_EPOCHS,
        imgsz=IMAGE_SIZE,
        batch=BATCH_SIZE,
        name=RUN_NAME,
        project=PROJECT_DIR # Specify the parent folder for results
    )

    # --- Confirmation that the model is saved in .pt format ---
    # Construct the full path to the saved best model weights (best.pt)
    # results.save_dir provides the path to the run directory (e.g., 'runs/detect/yolov8n_racket_detector')
    model_save_path = os.path.join(results.save_dir, 'weights', 'best.pt')

    # Load the best trained model to confirm it exists and is ready for use (optional step)
    best_model = YOLO(model_save_path)

    print("\n" + "="*50)
    print("Training Complete!")
    print("The final trained model is already saved in PyTorch (.pt) format.")
    print(f"The file path is:")
    print(f"--> {model_save_path}")
    print("="*50)

except Exception as e:
    print(f"A fatal error occurred during training: {e}")
    print("Check your CONFIG_FILE and ensure the dataset structure is correct.")

```

#Run the custom model on your test video

```python
# --- STEP 1: INSTALL REQUIRED LIBRARIES ---
# We need the official Ultralytics YOLO package, OpenCV for frame control, and 'tqdm' for a progress bar.
print("Step 1: Installing Ultralytics, OpenCV, and TQDM...")
!pip install ultralytics opencv-python tqdm -q
print("Installation complete.\n")

# --- STEP 2: FILE UPLOAD AND SETUP ---
# Import necessary Colab utilities and libraries
from google.colab import files
from IPython.display import HTML, display
from base64 import b64encode
import os
import time
import cv2
from ultralytics import YOLO
from tqdm.notebook import tqdm
```

You can choose to define the frame skip interval

This variable defines how many frames to skip before running the model again.

FRAME_SKIP = 1 means process every frame (no skipping).

FRAME_SKIP = 5 means process every 5th frame (1, 6, 11, 16, etc.)

Higher values increase speed but reduce tracking smoothness.

make sure to choose the right MODEL_FILENAME and VIDEO_FILENAME locations

```python
import pandas as pd
FRAME_SKIP = 1

# Define the expected filenames for clarity
MODEL_FILENAME = '/content/runs/detect/yolov8n_racket_detector/weights/best.pt' #'/content/bat_tracking_v2.pt'
VIDEO_FILENAME = '/content/baseball1.MP4'


# --- STEP 3: PERFORM OBJECT TRACKING with Frame Skipping ---
if os.path.exists(MODEL_FILENAME) and os.path.exists(VIDEO_FILENAME):
    print("Step 3: Starting Baseball Bat Tracking...")

    # Load the custom model
    try:
        model = YOLO(MODEL_FILENAME)
    except Exception as e:
        print(f"Error loading YOLO model: {e}")
        print("Please ensure your .pt file is a valid YOLOv8 model.")
        exit()

    # --- Custom Frame Processing Logic ---
    cap = cv2.VideoCapture(VIDEO_FILENAME)
    if not cap.isOpened():
        print(f"Error: Could not open video file {VIDEO_FILENAME}")
        exit()

    # Get video properties for VideoWriter
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')

    # Define output path more robustly
    output_project_folder = 'runs'
    output_tracking_subfolder = 'track_skipped_frames'
    # Extract base name of the video file (e.g., 'gopro3.MP4' from '/content/gopro3.MP4')
    base_video_name = os.path.basename(VIDEO_FILENAME)
    # Create a unique output directory for this run
    output_run_name = f"bat_tracking_results_skip_{FRAME_SKIP}"
    output_dir = os.path.join(output_project_folder, output_tracking_subfolder, output_run_name)
    os.makedirs(output_dir, exist_ok=True)
    # Construct the final output video path
    output_video_path = os.path.join(output_dir, f"tracked_{base_video_name}")

    # NOTE ON FPS: We keep the original FPS but only write the processed frames.
    # This will result in the output video playing back faster than the original
    # since frames are missing, but it accurately shows which frames were processed.
    writer = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

    current_frame_index = 0
    start_time = time.time()

    # Initialize list to store bounding box data
    bbox_data = []

    print(f"Video has {total_frames} frames. Processing...")

    # Use tqdm to show a nice progress bar across all frames
    for current_frame_index in tqdm(range(total_frames), desc="Tracking Progress"):
        success, frame = cap.read()
        if not success:
            break

        # --- CORE LOGIC: Check the skip interval ---
        if current_frame_index % FRAME_SKIP == 0:
            found_bbox_in_frame = False

            # Run the tracking only on the selected frame.
            # We use stream=True for efficiency.
            # The 'model.track' method with stream=True returns a generator.
            # We need to iterate through it to get the results.
            track_results_generator = model.track(
                source=frame, # Pass the frame data directly to the model
                tracker='botsort.yaml',
                conf= 0.1, #0.25, # You can try changing the Confidence based on your model performance
                iou= 0.3, #0.5, # Same for IoU
                stream=True,
                verbose=False
            )

            # Iterate through the generator to get the result object for the current frame
            for result in track_results_generator:
                # Get the annotated frame (result.plot() handles drawing boxes and tracks)
                annotated_frame = result.plot()

                # Write the annotated frame to the output video
                writer.write(annotated_frame)

                # Extract bounding box data if detections exist
                if result.boxes and len(result.boxes) > 0:
                    found_bbox_in_frame = True
                    for box in result.boxes:
                        x1, y1, x2, y2 = box.xyxy[0].tolist()
                        conf = box.conf[0].item()
                        cls = box.cls[0].item()
                        track_id = box.id[0].item() if box.id is not None and len(box.id) > 0 else -1 # Use -1 for no track ID

                        bbox_data.append({
                            'frame_num': current_frame_index,
                            'track_id': track_id,
                            'class_id': int(cls),
                            'confidence': round(conf, 4),
                            'x1': round(x1, 2),
                            'y1': round(y1, 2),
                            'x2': round(x2, 2),
                            'y2': round(y2, 2)
                        })

                # For a single frame source, we expect only one result, so break after the first.
                break

            # If no bounding boxes were found for the current processed frame, append NaN entries
            if not found_bbox_in_frame:
                bbox_data.append({
                    'frame_num': current_frame_index,
                    'track_id': float('nan'),
                    'class_id': float('nan'),
                    'confidence': float('nan'),
                    'x1': float('nan'),
                    'y1': float('nan'),
                    'x2': float('nan'),
                    'y2': float('nan')
                })

        # The tracker maintains the ID history even when frames are skipped,
        # so when the next processed frame comes, it tries to re-associate
        # the objects based on motion prediction (Kalman filter).

    # Cleanup
    cap.release()
    writer.release()

    end_time = time.time()
    print(f"\nTracking complete in {end_time - start_time:.2f} seconds.")
    print(f"Total frames processed: {current_frame_index // FRAME_SKIP}")
    print(f"Output video saved to disk.\n")

    # --- Save Bounding Box Data to CSV ---
    if bbox_data:
        bbox_df = pd.DataFrame(bbox_data)
        csv_filename = f"tracked_{os.path.splitext(base_video_name)[0]}_bboxes.csv"
        csv_output_path = os.path.join(output_dir, csv_filename)
        bbox_df.to_csv(csv_output_path, index=False)
        print(f"Bounding box data saved to: {csv_output_path}")

    else:
        print("No bounding box data to save.")


else:
    print("Required files were missing. Please re-run the cell and ensure both files are uploaded.")
```

#Optional plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# The CSV path specified by the user
csv_to_use = '/content/runs/track_skipped_frames/bat_tracking_results_skip_1/tracked_baseball_bboxes.csv'

# Load the bounding box data
bbox_df = pd.read_csv(csv_to_use)

# Calculate center coordinates
bbox_df['cx'] = (bbox_df['x1'] + bbox_df['x2']) / 2
bbox_df['cy'] = (bbox_df['y1'] + bbox_df['y2']) / 2

# Function to determine vector for the diagonal (Bottom-Left to Top-Right)
def get_diagonal_vector(row):
    vec_x_start = row['x1']  # Start from bottom-left x (same as top-left x)
    vec_y_start = row['y2']  # Start from bottom-left y
    vec_u = row['x2'] - row['x1']  # Horizontal component (x2 - x1)
    vec_v = row['y1'] - row['y2']  # Vertical component (y1 - y2) (from bottom y to top y)
    return pd.Series({'vec_x_start': vec_x_start, 'vec_y_start': vec_y_start, 'vec_u': vec_u, 'vec_v': vec_v})

# Apply the function to create vector columns for the diagonal
bbox_df[['vec_x_start', 'vec_y_start', 'vec_u', 'vec_v']] = bbox_df.apply(get_diagonal_vector, axis=1)

plt.figure(figsize=(14, 10))

# Plot bounding box centers
# Using 'alpha' to handle potential overplotting for many frames
plt.scatter(bbox_df['cx'], bbox_df['cy'], s=10, alpha=0.5, label='Bounding Box Center')

# Draw vectors for the diagonal
# Using 'scale_units='xy'' and 'angles='xy'' makes the vectors correctly scaled and oriented.
# 'scale=1' means the length of the arrow is equal to the length of the vector in data units.
# A small 'width' is used to make arrows visible but not too thick
plt.quiver(bbox_df['vec_x_start'], bbox_df['vec_y_start'],
           bbox_df['vec_u'], bbox_df['vec_v'],
           color='red', alpha=0.3, scale_units='xy', angles='xy', scale=1, width=0.002,
           label='Diagonal Vector (Bottom-Left to Top-Right)')

plt.xlabel('X Coordinate (pixels)')
plt.ylabel('Y Coordinate (pixels)')
plt.title(f'2D Plot of Bounding Box Centers and Diagonal Vectors (Bottom-Left to Top-Right) (Data from: {os.path.basename(csv_to_use)})')
plt.legend()
plt.grid(True)

# Set x and y limits to match video dimensions
# width and height variables are available from previous execution in the kernel state
plt.xlim(0, width)
plt.ylim(0, height) # Set y-axis limits from 0 to height

plt.gca().invert_yaxis()  # Invert y-axis to match typical image coordinates (0,0 at top-left)
plt.gca().set_aspect('equal', adjustable='box')  # Maintain aspect ratio for accurate representation
plt.show()
```
